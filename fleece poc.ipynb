```python
import openai

# Your OpenAI API key
openai.api_key = "sk-..."  # Replace with your actual key

# Define card reward rules for AMC (MCC 7832)
cards = {
    "Amex Gold": {
        "multiplier": 1,
        "notes": "No bonus on entertainment; best for groceries and dining."
    },
    "Chase Sapphire Preferred": {
        "multiplier": 2,
        "notes": "2x points on travel, which includes movie theaters."
    },
    "PayPal Debit": {
        "multiplier": 1,
        "notes": "Flat 1% cashback; no bonus categories."
    }
}

# Simulated AMC purchase
purchase_amount = 20

# Compute rewards
results = []
for card, details in cards.items():
    reward_value = purchase_amount * details["multiplier"]
    results.append((card, reward_value, details["notes"]))

# Sort by best reward
results.sort(key=lambda x: x[1], reverse=True)
top_card = results[0][0]

# Build summary string for GPT
summary = "\n".join(
    f"{card}: earns {reward:.2f} points/cashback â€” {note}"
    for card, reward, note in results
)

# Prompt GPT for recommendation explanation
gpt_prompt = f"""
I want to buy a $20 AMC movie ticket. Here are my cards and the rewards I get:

{summary}

Which card should I use and why?
"""

# Get GPT explanation
response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=[
        {"role": "user", "content": gpt_prompt}
    ],
    temperature=0.7
)

# Print result
print("Recommendation:\n")
print(response.choices[0].message.content)
```
